```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[20, 3, 7], [17, 6, 7]])
X
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[20, 3, 7], [17, 6, 7]])
X_pt
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflo
X_tf = tf.Variable([[20, 3, 7], [17, 6, 7]])
X_tf
```

```python
# lakukan perkalian matrix X, X_pt dan X_tf dengan suatu skalar
scalar = 3

X_scaled = X * scalar
X_pt_scaled = X_pt * scalar
X_tf_scaled = X_tf * scalar

print("Scaled NumPy matrix (X):")
print(X_scaled)

print("\nScaled PyTorch tensor (X_pt):")
print(X_pt_scaled)

print("\nScaled TensorFlow variable (X_tf):")
print(X_tf_scaled)
```

```python
# lakukan penjumlahan matrix X, X_pt dan X_tf dengan suatu skalar
scalar = 3

X_scaled = X + scalar
X_pt_scaled = X_pt + scalar
X_tf_scaled = X_tf + scalar

print("Scaled NumPy matrix (X):")
print(X_scaled)

print("\nScaled PyTorch tensor (X_pt):")
print(X_pt_scaled)

print("\nScaled TensorFlow variable (X_tf):")
print(X_tf_scaled)
```

```python
# lakukan pengurangan matrix X, X_pt dan X_tf dengan suatu skalar
scalar = 3

X_scaled = X - scalar
X_pt_scaled = X_pt - scalar
X_tf_scaled = X_tf - scalar

print("Scaled NumPy matrix (X):")
print(X_scaled)

print("\nScaled PyTorch tensor (X_pt):")
print(X_pt_scaled)

print("\nScaled TensorFlow variable (X_tf):")
print(X_tf_scaled)
```
