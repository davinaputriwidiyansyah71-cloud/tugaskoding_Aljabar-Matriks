# **Frobenius Norm**

1. menggunakan numpy, pytoch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm

menggunakan numpy

```python
import numpy as np
```

```python
# Membuat matriks 3x3
X = np.array([[26, 7, 11], [10, 4, 9],[6, 3, 1]])
X
```

```python
# Menghitung Frobenius Norm secara manual
(26**2 + 7**2 + 11**2 + 10**2 + 4**2 + 9**2 + 6**2 + 3**2 + 1**2)**(1/2)
```

```python
# Menghitung Frobenius Norm
np.linalg.norm(X)
```

menggunakan pytorch

```python
import torch
```

```python
#  Membuat matriks 3x3
X_pt = torch.tensor([[26, 7, 11], [10, 4, 9],[6, 3, 1]]) # torch.norm() supports floats only
X_pt
```

```python
# hitung nilai frobenius norm menggunakan Pytorch
np.linalg.norm(X) # same function as for vector L2 norm
```

menggunakan tensorflow

```python
import tensorflow as tf
```

```python
# Membuat matriks 3x3
X_tf = tf.Variable([[26, 7, 11], [10, 4, 9],[6, 3, 1]], dtype=tf.float32) # tf.norm() also supports floats only
X_tf
```

```python
# hitung nilai frobenius norm menggunakan TensorFlow

tf.norm(X_tf)
```
