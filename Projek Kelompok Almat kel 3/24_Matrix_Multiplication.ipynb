4



```python
# buat matrix B dengan ukuran 3x3 menggunakan NUMPY
B = np.array([[8, 10, 2],[9, 1, 3],[4, 0, 6]])
B
```

```python
# buat matrix B_pt dengan ukuran 3x3 menggunakan Pytorch
B_pt = torch.tensor([[8, 10, 2],[9, 1, 3],[4, 0, 6]])
B_pt
```

```python
# buat matrix B_tf dengan ukuran 3x3 menggunakan Tensorflow
B_tf = tf.constant([[8, 10, 2],[9, 1, 3],[4, 0, 6]])
B_tf

# Digunakan tf.constant karena nilai matriks/vektor bersifat tetap dan tidak mengalami perubahan selama program berjalan.
```

5

```python
# kalikan matrix A dengan matrix B (Numpy)
AB_np = np.dot(A, B)
AB_np
```

```python
# kalikan matrix A_pt dengan matrix B_pt (Pytorch)
AB_pt = torch.matmul(A_pt, B_pt)
AB_pt
```

```python
# kalikan matrix A_tf dengan matrix B_tf
AB_tf = tf.matmul(A_tf, B_tf)
AB_tf
```

6. buktikan apakah perkalian matrix bersifat komutatif? (AB=BA)?

```python
# NUMPY
BA_np = np.dot(B, A)   # ini akan ERROR

# karena (3x3) x (2x3)
# syarat perkalian matrix:
# jumlah kolom matrix pertama harus = jumlah baris matrix kedua
# di sini:
# kolom B = 3
# baris A = 2
# 3 ≠ 2 -> tidak memenuhi syarat -> maka akan terjadi ERROR
```

```python
# PYTORCH
BA_pt = torch.matmul(B_pt, A_pt)  # ERROR

# (3x3) x (2x3)
# kolom B (3) ≠ baris A (2)
# sehingga tidak bisa dikalikan
```

```python
# TENSORFLOW
BA_tf = tf.matmul(B_tf, A_tf)  # ERROR

# (3x3) x (2x3)
# kolom pertama (3) tidak sama dengan baris kedua (2)
# maka operasi tidak valid
```

**Kesimpulan**: Karena $BA$ bahkan tidak bisa dihitung dalam kasus ini, maka jelas $AB \neq BA$. Sekalipun kedua matriks berbentuk persegi (misal $3 \times 3$), hasil elemen di dalamnya biasanya akan berbeda. Jadi, perkalian matriks **tidak komutatif.**
