### Tensor Transposition

# 1

menggunakan numpy

```python
import numpy as np
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[9, 7, 26], [4, 8, 6]])
X
```

menggunakan Pytorch

```python
import torch
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[10, 47, 26], [0, 18, 86]])
X_pt
```

menggunakan TensorFlow

```python
import tensorflow as tf
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[11, 97, 46], [10, 68, 8]])
X_tf
```

# 2

```python
# lakukan transpose pada matrix X (Numpy)
X.T
```

```python
# lakukan transpose pada matrix X_pt (Pytorch)
X_pt.T
```

```python
# lakukan transpose pada matrix X_tf (TensorFlow)
tf.transpose(X_tf) # less Pythonic
```
