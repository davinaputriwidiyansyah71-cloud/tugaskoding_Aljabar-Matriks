Buat visualisasi dari persamaan berikut
y = 2x+1 dan 3x-y = 5



Substitution

$y = 2x + 1$

$3x - y = 5$

The second equation can be isolated to y :

$3x - y = 5$

$y = 5 - 3x$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
y1 = 2*x + 1
```

```python
y2 = 5 - 3*x
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 3])
ax.set_ylim([0, 8])
ax.plot(x, np.full_like(x, y1), c='green')
ax.plot(x, np.full_like(x, y2), c='brown')
plt.axvline(x=0.80, color='purple', linestyle='--')
_ = plt.axhline(y=2.55, color='purple', linestyle='--')
```

Elemination

$y=2x+1$

$3x−y=5$


---


Both equations can be rearranged to isolate  y . Starting with the first equation:

$y = 5 - 3x$


---
Then for the second equation:

$2x + y = 1$

$y = 1 - 2x$


```python
y1 = 5 - 3*x
```

```python
y2 = 1 - 2*x
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

# Add x and y axes:
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=4.10, color='purple', linestyle='--')
_ = plt.axhline(y=-7, color='purple', linestyle='--')
```
