y = 2x+1 dan 3x-y = 5



Substitution

$y = 2x + 1$

$3x - y = 5$

The second equation can be isolated to y :

$3x - y = 5$

$-y = 5 - 3x$

$y = 3x - 5$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
y1 = 2*x + 1
```

```python
y2 = 3*x - 5
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([1, 10])
ax.set_ylim([3, 15])
ax.plot(x, np.full_like(x, y1), c='green')
ax.plot(x, np.full_like(x, y2), c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
_ = plt.axhline(y=13, color='purple', linestyle='--')
```

Both equations can be rearranged to isolate  y . Starting with the first equation:

$-y = 5 - 3x$

$y = 3x - 5$
