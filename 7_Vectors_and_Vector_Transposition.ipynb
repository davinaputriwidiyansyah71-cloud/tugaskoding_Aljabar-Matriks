### Vectors (Rank 1 Tensors) in NumPy

```python
import numpy as np
```

```python
# Membuat vector x dengan ukuran 3x1
x = np.array([[26],[7],[11]]) # type argument is optional, e.g.: dtype=np.float16
x
```

```python
# Hitung panjang vector x
len(x)
```

```python
# Tampilkan ukuran vector x
x.shape
```

```python
# Tampilkan tipe data dari vector x
type(x)
```

```python
# Akses elemen pertama dari vector x
x[0] # zero-indexed
```

```python
# Tampilkan tipe data elemen pertama dari vector x
type(x[0,0])
```

*Vector Transposition*

```python
# Lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
x_t
```

```python
# Tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# Diasumsikan ada vector y
y = np.array([[26],[7],[11]])
y
```

```python
# tampilkan ukuran vektor y
y.shape
```

```python
# Lakukan transpose pada vector y dan simpan pada variabel y_t
y_t = y.T
y_t
```

```python
# Tampilkan ukuran dari vector y_t (koreksi soal: menampilkan y_t)
y_t.shape # this is a column vector as it has 3 rows and 1 column
```

```python
# Column vector can be transposed back to original row vector:
y_t.T
```

```python
y_t.T.shape
```
