```python
#buat matrix X ukuran 2x2 menggunakan numpy
import numpy as np
x = np.array([[1, 2], [3, 4]])
x
```

```python
#buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(x)
Xinv
```

```python
#kalikan X dengan Xinv menggunakan numpy
p = np.matmul(x, Xinv)
p
```

```python
#selesaikan persamaan berikut menggunakan numpy, tensorflow dan pytorch
import numpy as np

A = np.array([[4, 2], [-5, -3]])
B = np.array([4, -7])

x = np.linalg.solve(A, B)
x #hasil dari x merupakan nilai b dan c, dimana b adalah indeks pertama dan c adalah indeks kedua, sesuai dengan urutan di soal yaitu b pertama dan c kedua
```

```python
import torch

A_pt = torch.tensor([[4.0, 2.0], [-5.0, -3.0]])
B_pt = torch.tensor([4.0, -7.0])

x_pt = torch.linalg.solve(A_pt, B_pt)
x_pt # sama seperti NumPy tadi
```

```python
import tensorflow as tf

A_tf = tf.constant([[4.0, 2.0], [-5.0, -3.0]])
B_tf = tf.constant([[4.0], [-7.0]])

x_tf = tf.linalg.solve(A_tf, B_tf)
x_tf # sama seperti NumPy dan torch tadi untuk indeksnya
```
