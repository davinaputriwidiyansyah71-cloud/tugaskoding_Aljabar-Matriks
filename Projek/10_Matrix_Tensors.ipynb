**Matrics**

```python
import numpy as np
```

```python
# Membuat matrix X berukuran 2x3

X = np.array([[26, 7, 11], [20, 9, 16]])
X
```

```python
# Tampilkan ukuran dari matrix X
X.shape
```

```python
X.size
```

```python
# Tampilkan seluruh elemen dari kolom pertama

X[:,0]
```

```python
# Tampilkan seluruh elemen dari baris kedua

X[1,:]
```

### Matrices in **PyTorch**

```python
import torch
```

```python
# Membuat matrix X berukuran 2x3

X_pt = torch.tensor([[26, 7, 11], [20, 9, 16]])
X_pt
```

```python
# Tampilkan ukuran dari matrix X

X_pt.shape
```

```python
# Tampilkan seluruh elemen dari kolom pertama

X_pt[1,:]
```

**Matrices in TensorFlow**

```python
import tensorflow as tf
```

```python
# Membuat matrix X berukuran 2x3

X_tf = tf.Variable([[26, 7, 11], [20, 9, 16]])
X_tf
```

```python
tf.rank(X_tf)
```

```python
# Tampilkan ukuran dari matrix X

tf.shape(X_tf)
```

```python
# Tampilkan seluruh elemen dari kolom pertama

X_tf[1,:]
```
