# Penyelesaian Soal Transpose of **vector**

```python
import numpy as np
```

```python
x = np.array([[25], [2],  [-3], [-23]])
x
```

```python
# Tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# Lakukan transpose pada vector x
x_t = x.T
x_t
```

```python
# transposed back to original row vector
x_t.T
```
