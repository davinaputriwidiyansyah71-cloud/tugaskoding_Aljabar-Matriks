### Orthogonal Vectors

**1**

```python
import numpy as np
```

```python
# Membuat vector i
i = np.array([1, 0])
i
```

```python
# Membuat vector j
j = np.array([0, 1])
j
```

```python
# Menghitung dot product untuk membuktikan orthogonality

np.dot(i, j) # detail on the dot operation coming up...
```

**2**

```python
# membuat vector v
v = np.array([1, 2])
v
```

```python
# membuat  vector w
w = np.array([2, -1])
w
```

```python
# menghitung dot product membuktikan ortho
np.dot(v, w) # detail on the dot operation coming up...
```
