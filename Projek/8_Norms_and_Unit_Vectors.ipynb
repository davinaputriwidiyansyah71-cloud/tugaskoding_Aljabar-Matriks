### $L^2$ Norm

```python
import numpy as np
```

```python
# Membuat vector x dengan ukuran 3x1
x = np.array([[26], [7],[11]])
```

```python
x
```

```python
# Menghitung nilai L2 norm dari vector x
(26**2 + 7**2 + 11**2)**(1/2)
```

```python
# Menggunakan fungsi np.linalg.norm() yang merupakan standar di NumPy
np.linalg.norm(x)
```

 $L^1$ Norm

```python
# Membuat vector v dengan elemen (3,4)
v = np.array([3, 4])

print(f"Vector v: {v}")
```

```python
# Menghitung nilai L2 norm
norm = np.linalg.norm(v)

print(f"Nilai L2 norm v: {norm}")
```

```python
# Menampilkan unit vector
unit_v = v / norm

print(f"Unit vector v: {unit_v}")
```

```python
# Menghitung panjang unit vector

print("Panjang unit vector:", np.linalg.norm(unit_v))
```
